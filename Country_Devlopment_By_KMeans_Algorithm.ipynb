```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
country=pd.read_csv("Country_development_prediction.csv")
```

```python
#EDA
print(country.head())
print(country.columns)
print(country.isnull().sum())
print(country.shape)
```

```python
country.head()
```

```python
X=country.drop('country',axis=1)
Y=country['country']
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler  #scale krne k liye

X = StandardScaler().fit_transform(X)
```

```python
km=KMeans(n_clusters=2, init='k-means++',random_state=0)
km.fit(X)
y_kmeans=km.fit_predict(X)
centers=km.cluster_centers_
print(centers)
```

```python
plt.scatter(X[:,0], X[:,1], c=y_kmeans, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red')
plt.title("K-Means Clustering")
plt.show()
```
